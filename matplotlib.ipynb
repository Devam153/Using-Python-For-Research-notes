```python
x = np.linspace(0,10,21)
y = x**2
y1 = x**1.6
plt.plot(x,y, "bd-",linewidth = 3, markersize = 6, label = "first");
plt.plot(x,y1,"rs-", linewidth = 3, markersize = 6, label = "second");
plt.xlabel("$X$");
plt.ylabel("$Y$");
#plt.axis([xmin,xmax,ymin,ymax])
plt.axis([-0.5,10.5,-10,105]);
plt.legend(loc = "best");
#plt.savefig("myplot.pdf")
#print("file saved")
```

```python
#USING LOG
x = np.logspace(-1,1,40)
y = x**2
y1 = x**1.6
plt.loglog(x,y, "bd-",linewidth = 2, markersize = 4, label = "first,y");
plt.loglog(x,y1,"rs-", linewidth = 2, markersize = 4, label = "second, y1");
plt.xlabel("$X$");
plt.ylabel("$Y$");
#plt.axis([xmin,xmax,ymin,ymax])
#plt.axis([-0.5,10.5,-10,105]);
plt.legend(loc = "best");
print(x)

```
