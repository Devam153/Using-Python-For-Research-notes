```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
rolls = []
for i in range(1000000):
    rolls.append(random.choice([1,2,3,4,5,6]))
#print(rolls, end=' ')
```

```python
len(rolls)
```

```python
plt.hist(rolls, bins=np.linspace(0.5, 6.5, 7), edgecolor='black', color='skyblue')
```

```python
ys = []
for i in range(1000000):
    y = 0
    for j in range(10):
        x = random.choice([1,2,3,4,5,6])
        y += x
    ys.append(y)
#print(ys, end = '')
len(ys)
```

```python
min(ys)
```

```python
max(ys)
```

```python
plt.hist(ys, bins = np.linspace(10.5, 60.5, 51), edgecolor='black', color='lightgreen');
#added ; to only see the graph
```

But to understand what's happening here, we can again get some insights from probability theory. 
The so-called 'central limit theorem', or CLT, states that the sum of a large number of random variables 
regardless of their distribution will approximately follow a normal distribution.

but the main point is the following:

You can sum together many random variables whose distribution
is nothing like a normal distribution like die rolls, or even coin flips.
And yet, the distribution of the sum will get closer and closer
to a normal distribution as the number of random variables
that are added together increases.
